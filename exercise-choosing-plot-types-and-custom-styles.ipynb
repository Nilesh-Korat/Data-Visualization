```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Read the file into a variable spotify_data
spotify_data = pd.read_csv('spotify.csv', index_col="Date", parse_dates=True)
spotify_data.head()
```

# Try out seaborn styles

Run the command below to try out the `"dark"` theme.

```python
# Change the style of the figure
sns.set_style("dark")

# Line chart 
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```

```python
# Change the style of the figure to the "darkgrid" theme
sns.set_style("darkgrid")

# Line chart 
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```

Now, try out different themes by amending the first line of code and running the code cell again.  Remember the list of available themes:
- `"darkgrid"`
